```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Tanuj\Downloads\archive (3)\zomato.csv")
df.head()
```

```python
df.columns
```

```python
 df = df.drop(['url','address','phone','dish_liked','reviews_list','menu_item'],axis = 1)
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace = True)
df.shape
```

```python
df["rate"].unique()
```

```python
def handlerate(value):
    if(value== 'NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df.isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace = True)
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace =True)
df.head()
```

```python
df.info()
```

```python
df.rename(columns={'approx_cost(for two people)': 'cost2plate','listed_in(type)': 'type'},inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
df = df.drop(columns='listed_in(city)')
df.head()
```

```python
df['cost2plate'].unique()
```

```python
def handlecomma (value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)

df['cost2plate']=df['cost2plate'].apply(handlecomma)
df['cost2plate'].unique()
```

```python
df['rest_type'].value_counts()
```

```python
 rest_types = df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value

df['rest_type'] =df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
location =df['location'].value_counts(ascending = False)
```

```python
location_lessthan300 = location[location<300]
location_lessthan300
```

```python
def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value

df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
ax =sns.pairplot(df, hue = 'rest_type', palette= 'rainbow')
```

```python
df.head()
```

```python
cuisines = df['cuisines'].value_counts(ascending =False)
```

```python
cuisines_lessthan100 = cuisines[cuisines<100]
```

```python
def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
